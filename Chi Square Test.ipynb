# <font color = "#60BEE6"> Chi-Square Test Of Independence For This Scenario

# <font color = "#60BEE6"> 1. State The Hypotheses:

### <font color = "#60BEE6"> Null Hypothesis (H0): 
The type of smart home device purchased (Smart Thermostats vs. Smart Lights) and customer satisfaction level are independent.

### <font color = "#60BEE6"> Alternative Hypothesis (H1):
There is a significant association between the type of device purchased and customer satisfaction level.

# <font color = "#60BEE6"> 2. Data:
Replace the sample data with your actual data in the contingency table format.

```python
data_contingency=[[50,80,60,30,20],[70,100,90,50,50]]
```

### Compute the Chi-Square Statistic:

Well calculate the Chi-Square test statistic using the formula:X2=E(OE)2
where:

O) represents the observed frequency in each cell of the contingency table.

(E) represents the expected frequency in each cell (calculated based on row and column totals).

# <font color = "#60BEE6"> 3. Chi-Square Test:


```python
chi,p,df,expected = chi2_contingency(data_contingency)
```

 
The chi2_contingency function calculates the Chi-square statistic (chi2), p-value (pval), degrees of freedom (degrees_of_freedom), and expected frequencies (expected).

```python
print(f"chisquare value: {chi}\np value: {p}\ndegree of freedom: {df}\nexpected: {expected}")
```

# <font color = "#60BEE6"> 4. Make a Decision:

Compare the calculated Chi-Square statistic with the critical value:

If the p-value corresponding to the test statistic is less than 0.05, we reject the null hypothesis.
Otherwise, we fail to reject the null hypothesis.

### <font color = "#60BEE6"> Hypothesis Testing: Set the significance level (alpha) to 0.05.


```python
alpha=0.05
```

# <font color = "#60BEE6"> 5. Decision Rule:
Based on the p-value, we compare it to alpha. If p-value is less than or equal to alpha, we reject the null hypothesis (independence) and conclude a significant association. Otherwise, we fail to reject the null hypothesis.

```python
print(f"p value is {p}")
if p <= alpha:
    print("\nDecision: Reject the null hypothesis.")
    print("There is a significant association between device type and customer satisfaction level.")
else:
    print("\nDecision: Fail to reject the null hypothesis.")
    print("There is not enough evidence to conclude a significant association between device type and customer satisfaction level.")
```
